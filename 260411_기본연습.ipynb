```python
students = [
    {"name": "Ivy", "stat":90, "ds":80},
    {"name": "James", "stat":70, "ds":100 }
]
```

```python
def calculate_total(student):
    total = student["stat"] + student["ds"]
    student["total"]=total
```

```python
for s in students:
    calculate_total(s)

students
```

```python
wallets = [
    {"name": "생활비", "krw": 3400000},
    {"name": "학비", "krw": 17000000},
    {"name": "비상금", "krw": 850000}
]
```

```python
def convert_to_gbp(data):
    gbp = data["krw"]/1700
    data["gbp"] = gbp
```

```python
for w in wallets:
    convert_to_gbp(w)

for i, w in enumerate(wallets):
    print(f"{i+1}번 항목: {w["name"]}({w["gbp"]}GBP)")
```

```python
items = [
    {"name": "우유", "price": 1.2},
    {"name": "고기", "price": 15.5},
    {"name": "빵", "price": 2.5},
    {"name": "와인", "price": 25.0},
    {"name": "치즈", "price": 8.0}
]
```

```python
expensive_items = []
for i in items:
    if i["price"]>10:
        expensive_items.append(i)

expensive_items
```

```python
def apply_discount(item):
    if item["price"]>=10:
        item["price"] = item["price"]*0.9
        item["discount"] = "Yes"
    else:
        item["discount"] = "No"
```

```python
for i in items:
    apply_discount(i)

items
```

```python
total_revenue = 0
for i in items:
    total_revenue += i["price"]

total_revenue
```
